```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import export_graphviz
#from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_excel('/content/drive/MyDrive/Asssignment4_Data.xlsx',index_col='House ID')
df.head(5)
```

```python
column_list=[]
for i in df.columns:
  column_list.append(i)

del column_list[8]

column_list
```

```python
x=df[column_list]
y=df['Construction type']
print(x)
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```

###Changing the Depth =2

```python
clf2=DecisionTreeClassifier(max_depth=2, random_state=0)
clf2=clf2.fit(X_train,y_train)
y_pred=clf2.predict(X_test)
```

```python
pred=clf2.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(clf2,filled=True)
```

###Changing depth=5

```python
clf3=DecisionTreeClassifier(max_depth=5, random_state=0)
clf3=clf3.fit(X_train,y_train)
y_pred=clf3.predict(X_test)
```

```python
pred=clf3.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(clf3,filled=True)
```
